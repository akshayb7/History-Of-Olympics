```python
import plotly as py
import pandas as pd
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
%matplotlib inline

import os
os.listdir('input')

n = pd.read_csv('input/noc_regions.csv')
a = pd.read_csv('input/athlete_events.csv')

p = a.merge(n, on = 'NOC', how='left')
```

```python
mapping_df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv').drop(
    'GDP (BILLIONS)', axis=1)
def code_mapping():
    code_mapping = {}
    for i,j in mapping_df.values:
        code_mapping[i] = j

    # Individual mapping for countries with different names
    code_mapping['Democratic Republic of the Congo'] = 'COD'
    code_mapping['USA'] = 'USA'
    code_mapping['UK'] = 'GBR'
    code_mapping['Myanmar'] = 'MMR'
    code_mapping['Boliva'] = 'BOL'
    code_mapping['North Korea'] = 'PRK'
    code_mapping['South Korea'] = 'KOR'
    code_mapping['Ivory Coast'] = 'CIV'
    code_mapping['Republic of Congo'] = 'COG'
    code_mapping['Djibouti'] = 'DJI'
    return code_mapping
```

```python
def plot_map(df, col, title='Title', bar_title=''):
    # Get plotting CODES and add them to the data
    codes = code_mapping()
    df['CODE'] = df.region.map(codes)

    plot_df = pd.DataFrame(df.groupby('CODE')[col].count().reset_index())

    map_data = [dict(type = 'choropleth',
                     locations = plot_df.CODE,
                     z = plot_df[col],
                     autocolorscale =True, 
                     marker = dict(line = dict (color = 'rgb(180,180,180)', width = 0.5)),
                     colorbar = dict(autotick = True, tickprefix = '', title = bar_title))]
    layout = dict(title = title, 
                  geo = dict(showframe = True,
                             showcoastlines = False,
                             projection = dict(type = 'Mercator')))
    fig = dict(data=map_data, layout=layout )
    py.offline.iplot(fig, validate=False)
```

Greenland is an autonomous territory of Denmark and does not field its own Olympic team.

```python
plot_map(p,'Medal')
```
